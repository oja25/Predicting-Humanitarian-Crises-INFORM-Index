### 2015 Core Indicators Indeces

```python
# Select 20 specific indicator names
selected_indicator_names =  ['Access to electricity',
 'Conflict Barometer National Power',
 'Conflict Barometer Subnational',
 'Corruption Perception Index',
 'Frequency of Droughts events',
 'GDP per capita',
 'Gender Inequality Index',
 'Government Effectiveness',
 'Health expenditure per capita',
 'Highly Violent Conflict Probability',
 'IDP',
 'Internet users',
 'Literacy rate, adult total',
 'Mobile cellular subscriptions',
 'Palestinian Refugees',
 'Physicians density',
 'Population density (people per sq. km of land area)',
 'Refugees',
 'Returned Refugees',
 'Total population']  # Example list of indicator names to select

# Grouping by 'Indicator Name'
grouped_df1 = df1.groupby('IndicatorName')

# Creating a dictionary to store DataFrames
df1s = {}

# Splitting and storing DataFrames based on category
for category, group_df1 in grouped_df1:
    if category in selected_indicator_names:
        df1s[category] = group_df1.copy()

# Accessing the new DataFrames and printing info
for category, new_df1 in df1s.items():
    print(f"DataFrame of 2015 Core Indicators for Indicator {category}:")
    print(new_df1.head())
    print()  # line to space between dataframes
    print(new_df1.info())
    print()  # line to space between dataframes
```

```python
import os

# Define the folder path where you want to save the DataFrames
folder_path = '~/Desktop/CodeOp/DSF02/Group Project/2014-2024 Index by Indicator Id'  # Replace this with your desired folder path

# Accessing the new DataFrames and saving them
for category, new_df1 in df1s.items():
    # Construct the file path
    file_path = os.path.join(folder_path, f"2015 Select Core Indicators - {category}.csv")

    # Save each DataFrame as a separate CSV file in the specified folder
    new_df1.to_csv(file_path, index=False)
    print(f"DataFrame for 2015 Select Core Indicators for {category} saved as {file_path}")
```
