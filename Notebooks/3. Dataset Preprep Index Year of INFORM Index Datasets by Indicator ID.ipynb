### 2015 INFORM Indeces

```python
# Select 11 specific indicator IDs
selected_indicator_ids = ['INFORM', 'HA', 'HA.NUM', 'HA.NUM.CON', 'HA.NAT.EPI', 'VU', 'VU.SEV', 'VU.VGR', 'CC', 'CC.INF', 'CC.INS.GOV']  # Example list of indicator IDs to select

# Grouping by 'Indicator ID'
grouped_df1 = df1.groupby('IndicatorId')

# Creating a dictionary to store DataFrames
df1s = {}

# Splitting and storing DataFrames based on category
for category, group_df1 in grouped_df1:
    if category in selected_indicator_ids:
        df1s[category] = group_df1.copy()

# Accessing the new DataFrames and printing info
for category, new_df1 in df1s.items():
    print(f"DataFrame of 2015 Indices for Indicator Id {category}:")
    print(new_df1.head())
    print()  # line to space between dataframes
    print(new_df1.info())
    print()  # line to space between dataframes
```

```python
import os

# Define the folder path where you want to save the DataFrames
folder_path = '~/Desktop/CodeOp/DSF02/Group Project/2014-2024 Index by Indicator Id'  # Replace this with your desired folder path

# Accessing the new DataFrames and saving them
for category, new_df1 in df1s.items():
    # Construct the file path
    file_path = os.path.join(folder_path, f"2015 Select INFORM Index - {category}.csv")

    # Save each DataFrame as a separate CSV file in the specified folder
    new_df1.to_csv(file_path, index=False)
    print(f"DataFrame for 2015 Select INFORM Year Indeces for {category} saved as {file_path}")
```
